```python
# Importando las bibliotecas necesarias para visualización de datos y análisis exploratorio
import matplotlib.pyplot as plt  # Biblioteca para crear gráficos y visualizaciones estáticas, animadas e interactivas en Python
import seaborn as sns  # Biblioteca para crear gráficos estadísticos atractivos y informativos en Python
import pandas as pd # Importar la biblioteca pandas con el alias pd
import numpy as np # Importando la biblioteca numpy con el alias np
```

```python
# Leer datos desde archivo CSV
csv_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSWVPXE4NhZdrxbsSoFt0gI_kTvmKkR1qLmiOhkOcvev0htnMyJa3sgQHNQg6zna5fSHOye24Nl5M3N/pub?gid=1749908058&single=true&output=csv"
df = pd.read_csv(csv_path, sep=",")
df
```

```python
# Utiliza el método isnull() para obtener un DataFrame de valores booleanos
valores_nulos = df.isnull().sum()

# Imprime el total de valores nulos por columna
print(valores_nulos)
```

```python
#Elimina datos nulos presentes en el DataFrame
df.dropna(inplace=True)
```

```python
# Utiliza el método isnull() para obtener un DataFrame de valores booleanos
valores_nulos = df.isnull().sum()

# Imprime el total de valores nulos por columna
print(valores_nulos)
```

```python
# Convierte la columna 'fecha' en un objeto de fecha
df['date'] = pd.to_datetime(df['date'])

# Crea nuevas columnas 'año' y 'mes' a partir de la columna 'fecha'
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Muestra el DataFrame resultante
print(df)
```

```python
# Elimina la columna 'date' usando el método drop
df = df.drop(columns=['date'])

# Muestra el DataFrame resultante sin la columna 'date'
print(df)
```

```python
# Reordenar las columnas manualmente
df = df[['price','mileage','power','make_model','year','month','fuel','gear','offer_type','version','link']]

# Muestra el DataFrame resultante
print(df)
```
