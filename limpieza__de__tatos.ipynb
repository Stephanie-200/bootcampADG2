```python
# Importando las bibliotecas necesarias para visualización de datos y análisis exploratorio
import matplotlib.pyplot as plt  # Biblioteca para crear gráficos y visualizaciones estáticas, animadas e interactivas en Python
import seaborn as sns  # Biblioteca para crear gráficos estadísticos atractivos y informativos en Python
import pandas as pd # Importar la biblioteca pandas con el alias pd
import numpy as np
```

```python
# Leer datos desde archivo CSV
csv_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQdyI3u0qboOC-apHnWALG5znxZo7y13yLHuvZhArPdpeahesK3QiVcoa-mIleD4Cggz81jCrDM2Cfo/pub?output=csv"
df = pd.read_csv(csv_path, sep=",",header=None)

df
```

```python
#agregamos la función isnull.
df.isnull()
```

```python
# Supongamos que tienes un DataFrame llamado 'df'
# Utiliza el método isnull() para obtener un DataFrame de valores booleanos
valores_nulos = df.isnull()

# Utiliza el método sum() en el DataFrame de valores booleanos para contar los valores nulos en cada columna
total_nulos_por_columna = valores_nulos.sum()

# Imprime el total de valores nulos por columna
print(total_nulos_por_columna)
```

```python
#elimina las filas que contienen datos nulos
df.dropna()
```
